```python
def load_img_uint8(path):
    img = pyplot.imread(path)
    if img.ndim==3 and img.shape[2] == 4:
        img = img[:, :, :3].copy()
    if img.dtype != np.uint8:
        img = np.uint8(img*255)
        
    return img
    
```

```python
slika_A = load_img_uint8('primer_preprost_1.png')[::2, ::2]
slika_B = load_img_uint8('primer_preprost_2.png')[::2, ::2]
```

## Poravnava brez barvne informacije

Pričakuje se poravnava krogov z napačnimi barvnimi pari.

```python
R, tr = poravnava.oceni_transformacijo_ICP(slika_A, slika_B, max_itt=30)
slika_A_tr = poravnava.transformiraj_sliko(slika_A, R, tr)

fig, ax_list = pyplot.subplots(1, 3)
ax_list[0].imshow(slika_A)
ax_list[0].set_title('slika A')
ax_list[1].imshow(slika_B)
ax_list[1].set_title('slika B')
ax_list[2].imshow(slika_A_tr)
ax_list[2].set_title('slika A poravnava z B')

for ax in ax_list:
    ax.set_axis_off()
```

## Poravnava z uporabo barv

Pričakuje se pravilna poravnava krogov tudi po barvah.

```python
R, tr = poravnava.oceni_transformacijo_ICP(slika_A, slika_B, max_itt=30, barve=True)
slika_A_tr = poravnava.transformiraj_sliko(slika_A, R, tr)

fig, ax_list = pyplot.subplots(1, 3)
ax_list[0].imshow(slika_A)
ax_list[0].set_title('slika A')
ax_list[1].imshow(slika_B)
ax_list[1].set_title('slika B')
ax_list[2].imshow(slika_A_tr)
ax_list[2].set_title('slika A poravnava z B')

for ax in ax_list:
    ax.set_axis_off()
```
